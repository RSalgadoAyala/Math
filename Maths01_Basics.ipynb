```python
def derivative_x(f, x, step_size):
  m = (f(x + step_size) - f(x)) / ((x + step_size) - x)
  return m
```

```python
def my_function(x):
  return x**2
```

```python
slope_at_2 = derivative_x(my_function, 2, 0.00001)
print(slope_at_2)
```

Using limits to calculate a slope

```python
x, s = symbols('x s')

f= x**2

slope_f = (f.subs(x, x+s) - f) / ((x+s) - x)

slope_2 = slope_f.subs(x, 2)

result = limit(slope_2, s, 0)

print(result)
```

Chain rule

```python
z = (x**2 + 1)**3 - 2
dz_dx = diff(z, x)
print(dz_dx)
```

```python
x, y = symbols(' x y')

_y = x**2 + 1
dy_dx = diff(_y)

z = y**3 - 2
dz_dy = diff(z)

dz_dx_chain = (dy_dx * dz_dy).subs(y, _y)
dz_dx_no_chain = diff(z.subs(y, _y))

print(dz_dx_chain)
print(dz_dx_no_chain)
```

Integrals

```python
def approximate_integral(a, b, n, f):
  delta_x = (b - a) / n
  total_sum = 0

  for i in range(1, n + 1):
    midpoint = 0.5 * (2 * a + delta_x * (2*i-1))
    total_sum += f(midpoint)

  return total_sum * delta_x

def my_function(x):
  return x**2 + 1
```

```python
area = approximate_integral(a=0, b=1, n=10, f=my_function)

print(area)
```

```python
x = symbols('x')

f = x**2 + 1

# Calculate the integral of the function with respect to x
# for the area between x = 0 and 1

area = integrate(f, (x, 0, 1))

print(area)
```

Using limits to calculate integrals

```python
# Declare variables to SymPy
x , i, n = symbols('x i n')

# Declare function and range
f = x**2 + 1
lower, upper = 0, 1

# Calculate widht and each rectangle height at index "i"
delta_x = ((upper - lower) / n)
x_i = (lower + delta_x * i)
fx_i= f.subs(x, x_i)

# Iterate all "n" rectangles and sum their areas
n_rectangles = Sum(delta_x * fx_i, (i, 1, n)).doit()

# Calculate the ara by approaching the number of rectangles "n" to infinity
area = limit(n_rectangles, n, oo)

print(area)
```
